````python
# RECURSIVE:

import re

def mermaid_process(text):

    result = re.findall('```mermaid(?s:.*?)```', text, re.M)

    print(len(result))

    if len(result) <= 0:

        return text

    sub_text = result[0][len('```mermaid'):-3]

    # print(sub_text)

    text = re.sub('```mermaid(?s:.*?)```', '<div class="mermaid">'+ sub_text + '</div>', text, 1)

    print(text)

    return mermaid_process(text)
````
